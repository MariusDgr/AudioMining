```python
import matplotlib.pyplot as plt
from scipy import signal
from scipy.io import wavfile
```

```python
# Get data files

import os 
two_up =  os.path.abspath(os.path.join('.' ,"../.."))
print("Project root path is: ", two_up)

className = 'violin'
data_path = os.path.join(two_up, 'data', 'raw', 'mp3', className)

root_paths = []

files = []
# r=root, d=directories, f = files
for r, d, f in os.walk(data_path):
    for file in f:
        if '.mp3' in file:
            files.append(os.path.join(r, file))

for f in files:
    print(f)
    break
    
```

```python
# Transform mp3 to wav
from pydub import AudioSegment

os.makedirs(os.path.join(two_up, "data", "converted_wav", className), exist_ok=True)

for f in files:
    pSplit = os.path.split(f)
    fileName = pSplit[1]
    className = os.path.split(pSplit[0])[1]
        
    sound = AudioSegment.from_mp3(f)
    wavPath = os.path.join(two_up, 'data', 'converted_wav', className, fileName.replace(".mp3", ""))
    sound.export(wavPath+'.wav', format="wav")
    break
```
